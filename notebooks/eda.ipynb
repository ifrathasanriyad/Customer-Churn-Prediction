```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw_data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ১. থিম সেট করা এবং একটি বড় ক্যানভাস (Figure) ও ১২টি সাব-প্লট (Axes) তৈরি করা
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(22, 26))

# ছবির মেইন টাইটেল বা শিরোনাম
fig.suptitle('Customer Demographics & Services vs Churn Analysis', fontsize=20, fontweight='bold', y=0.96)


# --- প্লট ১: Overall Churn Distribution (Row 0, Col 0) ---
ax1 = sns.countplot(data=df, x='Churn', hue='Churn', palette='Set2', ax=axes[0, 0], legend=False)
for container in ax1.containers:
    ax1.bar_label(container)
ax1.set_title('Overall Churn Distribution', fontsize=12, fontweight='bold')
ax1.set_xlabel('Churn')
ax1.set_ylabel('Number of Customers')


# --- প্লট ২: Churn Count by Gender (Row 0, Col 1) ---
ax2 = sns.countplot(data=df, x='gender', hue='Churn', palette='Set1', ax=axes[0, 1])
for container in ax2.containers:
    ax2.bar_label(container)
ax2.set_title('Churn Count by Gender', fontsize=12, fontweight='bold')
ax2.set_xlabel('Gender')
ax2.set_ylabel('Number of Customers')
ax2.legend(title='Churn Status', labels=['No Churn (0)', 'Churn (1)'])


# --- প্লট ৩: Churn by Senior Citizen (Row 0, Col 2) ---
ax3 = sns.countplot(data=df, x='SeniorCitizen', hue='Churn', palette='Set2', ax=axes[0, 2])
for container in ax3.containers:
    ax3.bar_label(container)
ax3.set_title('Senior Citizens Churn Comparison', fontsize=12, fontweight='bold')
ax3.set_xlabel('')
ax3.set_ylabel('Number of Customers')
ax3.set_xticks([0, 1])
ax3.set_xticklabels(['Non-Senior Citizen (0)', 'Senior Citizen (1)'])
ax3.legend(title='Churn Status', labels=['No Churn (0)', 'Churn (1)'])


# --- প্লট ৪: Phone Service vs Churn (Row 1, Col 0) ---
ax4 = sns.countplot(data=df, x='PhoneService', hue='Churn', palette='Set1', ax=axes[1, 0])
for container in ax4.containers:
    ax4.bar_label(container)
ax4.set_title('Phone Service vs Churn', fontsize=12, fontweight='bold')
ax4.set_xlabel('Phone Service')
ax4.set_ylabel('Number of Customers')
ax4.legend(title='Churn Status', labels=['No Churn (0)', 'Churn (1)'])


# --- প্লট ৫: Internet Service vs Churn (Row 1, Col 1) ---
ax5 = sns.countplot(data=df, x='InternetService', hue='Churn', palette='Set2', ax=axes[1, 1])
for container in ax5.containers:
    ax5.bar_label(container)
ax5.set_title('Internet Service vs Churn', fontsize=12, fontweight='bold')
ax5.set_xlabel('Internet Service Type')
ax5.set_ylabel('Number of Customers')
ax5.legend(title='Churn Status', labels=['No Churn (0)', 'Churn (1)'])


# --- প্লট ৬: Online Security vs Churn (Row 1, Col 2) ---
ax6 = sns.countplot(data=df, x='OnlineSecurity', hue='Churn', palette='Set3', ax=axes[1, 2])
for container in ax6.containers:
    ax6.bar_label(container)
ax6.set_title('Online Security vs Churn', fontsize=12, fontweight='bold')
ax6.set_xlabel('Online Security')
ax6.set_ylabel('Number of Customers')
ax6.legend(title='Churn Status', labels=['No Churn (0)', 'Churn (1)'])


# --- --- প্লট ৭: Online Backup vs Churn (Row 2, Col 0) ---
ax7 = sns.countplot(data=df, x='OnlineBackup', hue='Churn', palette='Set1', ax=axes[2, 0])
for container in ax7.containers:
    ax7.bar_label(container)
ax7.set_title('Online Backup vs Churn', fontsize=12, fontweight='bold')
ax7.set_xlabel('Online Backup')
ax7.set_ylabel('Number of Customers')
ax7.legend(title='Churn Status', labels=['No Churn (0)', 'Churn (1)'])


# --- প্লট ৮: Device Protection vs Churn (Row 2, Col 1) ---
ax8 = sns.countplot(data=df, x='DeviceProtection', hue='Churn', palette='Set2', ax=axes[2, 1])
for container in ax8.containers:
    ax8.bar_label(container)
ax8.set_title('Device Protection vs Churn', fontsize=12, fontweight='bold')
ax8.set_xlabel('Device Protection')
ax8.set_ylabel('Number of Customers')
ax8.legend(title='Churn Status', labels=['No Churn (0)', 'Churn (1)'])


# --- প্লট ৯: Tech Support vs Churn (Row 2, Col 2) ---
ax9 = sns.countplot(data=df, x='TechSupport', hue='Churn', palette='Set3', ax=axes[2, 2])
for container in ax9.containers:
    ax9.bar_label(container)
ax9.set_title('Tech Support vs Churn', fontsize=12, fontweight='bold')
ax9.set_xlabel('Tech Support')
ax9.set_ylabel('Number of Customers')
ax9.legend(title='Churn Status', labels=['No Churn (0)', 'Churn (1)'])


# --- প্লট ১০: Contract vs Churn (Row 3, Col 0) ---
ax10 = sns.countplot(data=df, x='Contract', hue='Churn', palette='Set1', ax=axes[3, 0])
for container in ax10.containers:
    ax10.bar_label(container)
ax10.set_title('Contract vs Churn', fontsize=12, fontweight='bold')
ax10.set_xlabel('Contract Type')
ax10.set_ylabel('Number of Customers')
ax10.legend(title='Churn Status', labels=['No Churn (0)', 'Churn (1)'])


# --- প্লট ১১: Paperless Billing vs Churn (Row 3, Col 1) ---
ax11 = sns.countplot(data=df, x='PaperlessBilling', hue='Churn', palette='Set2', ax=axes[3, 1])
for container in ax11.containers:
    ax11.bar_label(container)
ax11.set_title('Paperless Billing vs Churn', fontsize=12, fontweight='bold')
ax11.set_xlabel('Paperless Billing')
ax11.set_ylabel('Number of Customers')
ax11.legend(title='Churn Status', labels=['No Churn (0)', 'Churn (1)'])


# --- প্লট ১২: Payment Method vs Churn (Row 3, Col 2) ---
ax12 = sns.countplot(data=df, x='PaymentMethod', hue='Churn', palette='Set3', ax=axes[3, 2])
for container in ax12.containers:
    ax12.bar_label(container)
ax12.set_title('Payment Method vs Churn', fontsize=12, fontweight='bold')
ax12.set_xlabel('Payment Method')
ax12.set_ylabel('Number of Customers')
# Payment Method এর নামগুলো একটু বড় হওয়ায় লেখাগুলোকে ১৫ ডিগ্রি বাঁকা (rotate) করা হয়েছে
ax12.tick_params(axis='x', rotation=15)
ax12.legend(title='Churn Status', labels=['No Churn (0)', 'Churn (1)'])


# সাব-প্লটগুলোর মাঝে ফাঁকা জায়গা সুন্দরভাবে ম্যানেজ করা
plt.tight_layout(rect=[0, 0, 1, 0.94])

# সম্পূর্ণ চিত্রটি প্রদর্শন করা
plt.show()
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

# hue='Churn' এবং multiple='stack' দিয়ে চমৎকার একটি স্ট্যাকড হিস্টোগ্রাম তৈরি
sns.histplot(data=df, x='tenure', hue='Churn', multiple='stack', kde=True, bins=30, palette='Set1')

plt.title('Tenure Distribution by Churn Status', fontsize=14, pad=15, fontweight='bold')
plt.xlabel('Tenure (Months)', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.legend(title='Churn Status', labels=['Churn (1)', 'No Churn (0)'])

plt.show()
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.kdeplot(data=df, x='tenure', hue='Churn', fill=True, common_norm=False, palette='Set1', alpha=0.5)

plt.title('KDE Plot: Tenure Distribution by Churn Status', fontsize=14, pad=15, fontweight='bold')
plt.xlabel('Tenure (Months)', fontsize=12)
plt.ylabel('Density', fontsize=12)

plt.legend(title='Churn Status', labels=['Churn (1)', 'No Churn (0)'])

plt.show()
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

ax = sns.boxplot(data=df, x='Churn', y='tenure', palette='Set2', width=0.5)

plt.title('Box Plot: Tenure Distribution by Churn Status', fontsize=14, pad=15, fontweight='bold')
plt.xlabel('Churn Status', fontsize=12)
plt.ylabel('Tenure (Months)', fontsize=12)

plt.xticks(ticks=[0, 1], labels=['No Churn (0)', 'Churn (1)'])

plt.show()
```

Tenure vs Churn Analysis

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))

df['Churn_numeric'] = df['Churn'].map({'Yes': 1, 'No': 0, 1: 1, 0: 0})

churn_rate_by_tenure = df.groupby('tenure')['Churn_numeric'].mean() * 100

sns.lineplot(x=churn_rate_by_tenure.index, y=churn_rate_by_tenure.values, 
             color='crimson', linewidth=2.5, marker='o', markersize=5)

plt.title('Churn Rate (%) over Customer Tenure (Months)', fontsize=14, pad=15, fontweight='bold')
plt.xlabel('Tenure (Months)', fontsize=12)
plt.ylabel('Churn Rate (%)', fontsize=12)
plt.ylim(0, 100)  

plt.show()
```
