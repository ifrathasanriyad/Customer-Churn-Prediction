Monthly Charges: KDE Plot

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

# common_norm=False ব্যবহার করা হয়েছে যাতে প্রতিটি গ্রুপ তার নিজস্ব স্কেলে ডেনসিটি দেখায়
sns.kdeplot(data=df, x='MonthlyCharges', hue='Churn', fill=True, common_norm=False, palette='Set1', alpha=0.5)

plt.title('KDE Plot: Monthly Charges Distribution by Churn Status', fontsize=14, pad=15, fontweight='bold')
plt.xlabel('Monthly Charges ($)', fontsize=12)
plt.ylabel('Density', fontsize=12)

plt.legend(title='Churn Status', labels=['Churn (1)', 'No Churn (0)'])

plt.show()
```

The KDE plot reveals that customers paying around $20 per month represent a large, highly stable user base with an extremely low churn rate, likely due to the affordability of basic plans. Conversely, there is a sharp increase in churn within the premium $70 to $100 price bracket. This high-paying segment often bundles multiple expensive features—such as fiber-optic internet, online security, and device protection—making them highly price-sensitive and prone to switching to competitors offering better deals. Consequently, retention strategies and targeted loyalty discounts should prioritize this $70+ billing group to mitigate high-value subscriber loss.

Monthly Charges: Box Plot

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.boxplot(data=df, x='Churn', y='MonthlyCharges', hue='Churn', palette='Set2', width=0.5, legend=False)

plt.title('Box Plot: Monthly Charges Distribution by Churn Status', fontsize=14, pad=15, fontweight='bold')
plt.xlabel('Churn Status', fontsize=12)
plt.ylabel('Monthly Charges ($)', fontsize=12)

plt.xticks(ticks=[0, 1], labels=['No Churn (0)', 'Churn (1)'])

plt.show()
```

The box plot shows a clear difference in monthly bills between customers who stayed and those who left. The median bill for churned customers is much higher, sitting around $80, while active customers typically pay less, with a median near $65. In fact, most of the customers who left were clustered in the higher pricing bracket. This proves that expensive plans are directly pushing people to cancel their service, meaning the company needs to offer better deals or discounts to high-paying users to keep them from leaving.

The KDE plot shows a clear difference in total spending between customers who stayed and those who left. Most of the customers who cancelled their service are heavily clustered at the very beginning of the curve, showing that they drop out after spending very little money. On the other hand, as total charges increase, the churn curve flattens out completely while the active customer curve stretches far to the right. This proves that once customers cross a certain spending threshold and invest more over time, they become highly loyal and rarely leave the company.

Correlation Analysis

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

numerical_df = df.select_dtypes(include=['int64', 'float64'])

corr_matrix = numerical_df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)

plt.title('Correlation Heatmap of Numerical Features', fontsize=14, pad=15, fontweight='bold')

plt.show()
```

The heatmap shows a clear connection between our numerical features and customer churn. The strongest link is between tenure and total charges (0.83), proving that long-term customers naturally spend more money over time. When looking at churn, it has a clear negative connection with tenure (-0.35), which proves that the longer a customer stays, the less likely they are to leave. On the other hand, churn has a positive link with monthly charges (0.19), confirming that higher bills are actively pushing people to cancel their service.

Relationship Analysis: Pair Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ০. TotalCharges-কে নিউমেরিক্যালে রূপান্তর করা ও খালি ঘর বাদ দেওয়া
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
plot_df = df.dropna(subset=['TotalCharges', 'tenure', 'MonthlyCharges', 'Churn'])

# ১. থিম সেট করা
sns.set_theme(style="ticks")

# ২. Pair Plot তৈরি করা
# x-কে hue হিসেবে দেওয়ার নতুন নিয়মের সাথে সামঞ্জস্য রেখে
g = sns.pairplot(data=plot_df, vars=['tenure', 'MonthlyCharges', 'TotalCharges'], 
                 hue='Churn', palette='Set1', diag_kind='kde', plot_kws={'alpha': 0.6})

# ৩. টাইটেল সেট করা
g.fig.suptitle('Pair Plot of Numerical Features by Churn Status', fontsize=14, fontweight='bold', y=1.02)

# গ্রাফ প্রদর্শন করা
plt.show()
```

The pair plot gives us a complete look at how our numerical features interact based on churn. It clearly shows that customers who left (marked in blue) are heavily clustered in the zone where tenure is very low but monthly charges are quite high. We can also see a sharp, triangle-like relationship between tenure and total charges, showing that long-term customers naturally pile up higher total bills. Overall, the plot confirms that the highest risk of losing customers happens right at the start of their journey if their monthly bills are too expensive.

Outlier Detection

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# ২. Tenure-এর জন্য Box Plot
sns.boxplot(data=df, x='Churn', y='tenure', hue='Churn', palette='Set2', width=0.4, ax=axes[0], legend=False)
axes[0].set_title('Tenure Distribution', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Churn Status')
axes[0].set_ylabel('Tenure (Months)')
axes[0].set_xticklabels(['No Churn (0)', 'Churn (1)'])

# ৩. Monthly Charges-এর জন্য Box Plot
sns.boxplot(data=df, x='Churn', y='MonthlyCharges', hue='Churn', palette='Set2', width=0.4, ax=axes[1], legend=False)
axes[1].set_title('Monthly Charges Distribution', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Churn Status')
axes[1].set_ylabel('Monthly Charges ($)')
axes[1].set_xticklabels(['No Churn (0)', 'Churn (1)'])

# ৪. Total Charges-এর জন্য Box Plot
sns.boxplot(data=df, x='Churn', y='TotalCharges', hue='Churn', palette='Set2', width=0.4, ax=axes[2], legend=False)
axes[2].set_title('Total Charges Distribution', fontsize=12, fontweight='bold')
axes[2].set_xlabel('Churn Status')
axes[2].set_ylabel('Total Charges ($)')
axes[2].set_xticklabels(['No Churn (0)', 'Churn (1)'])

plt.tight_layout()
plt.show()
```

The box plots show a clear look at our data distribution and help us spot any unusual points. While monthly charges look completely clean, we can see some isolated dots at the top of the churn boxes for both tenure and total charges. These dots represent outliers—specifically, a few rare long-term customers with high total bills who still decided to leave. However, since these points are tightly grouped right above the whiskers, they are safe to keep and show that our data is in great shape for modeling.
